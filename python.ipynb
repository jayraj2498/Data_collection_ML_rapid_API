```python
import pandas as pd
import ast

# Define a function to safely convert string to dictionary
def safe_literal_eval(x):
    try:
        return ast.literal_eval(x)
    except (ValueError, SyntaxError):
        return {}

# Assuming 'df' is your DataFrame
# Convert 'cases' column from string to dictionary safely
df['cases'] = df['cases'].apply(safe_literal_eval)

# Extract relevant information into separate columns
df['new_cases'] = df['cases'].apply(lambda x: x.get('new'))
df['active_cases'] = df['cases'].apply(lambda x: x.get('active'))
df['critical_cases'] = df['cases'].apply(lambda x: x.get('critical'))
df['recovered_cases'] = df['cases'].apply(lambda x: x.get('recovered'))
df['total_cases'] = df['cases'].apply(lambda x: x.get('total'))
df['cases_per_million'] = df['cases'].apply(lambda x: x.get('1M_pop'))

# Drop the original 'cases' column
df.drop('cases', axis=1, inplace=True)

# Display the DataFrame with the new columns
print(df.head())
```

```python
import requests

url = "https://bloomberg-market-and-financial-news.p.rapidapi.com/market/auto-complete"
querystring = {"query": "<REQUIRED>"}
headers = {
    "X-RapidAPI-Key": "1e559a1c55mshee1e34f721c08f3p10dc50jsnbbdfbe15b9aa",
    "X-RapidAPI-Host": "bloomberg-market-and-financial-news.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)
data = response.json()

# Extract and print quotes
print("Quotes:")
for quote in data.get('quote', []):
    print(f"Symbol: {quote.get('symbol')}")
    print(f"Name: {quote.get('name')}")
    print(f"Region: {quote.get('region')}")
    print(f"Country: {quote.get('country')}")
    print(f"Resource Type: {quote.get('resourceType')}")
    print()

# Extract and print news articles
print("News Articles:")
for news in data.get('news', []):
    print(f"Title: {news.get('title')}")
    print(f"Date: {news.get('date')}")
    print(f"URL: {news.get('longURL')}")
    print()

```

```python
from tabulate import tabulate

# Extract quotes into a list of lists
quotes_table = []
for quote in data.get('quote', []):
    quotes_table.append([
        quote.get('symbol'),
        quote.get('name'),
        quote.get('region'),
        quote.get('country'),
        quote.get('resourceType')
    ])

# Extract news articles into a list of lists
news_table = []
for news in data.get('news', []):
    news_table.append([
        news.get('title'),
        news.get('date'),
        news.get('longURL')
    ])

# Print quotes table
print("Quotes:")
print(tabulate(quotes_table, headers=["Symbol", "Name", "Region", "Country", "Resource Type"], tablefmt="grid"))

# Print news articles table
print("\nNews Articles:")
print(tabulate(news_table, headers=["Title", "Date", "URL"], tablefmt="grid"))
```

```python
import pandas as pd

# Extract quotes into a DataFrame
quotes_data = []
for quote in data.get('quote', []):
    quotes_data.append({
        'Symbol': quote.get('symbol'),
        'Name': quote.get('name'),
        'Region': quote.get('region'),
        'Country': quote.get('country'),
        'Resource Type': quote.get('resourceType')
    })

quotes_df = pd.DataFrame(quotes_data)

# Extract news articles into a DataFrame
news_data = []
for news in data.get('news', []):
    news_data.append({
        'Title': news.get('title'),
        'Date': pd.to_datetime(news.get('date'), unit='s'),  # Convert Unix timestamp to datetime
        'URL': news.get('longURL')
    })

news_df = pd.DataFrame(news_data)

# Display quotes DataFrame
print("Quotes:")
print(quotes_df)

# Display news articles DataFrame
print("\nNews Articles:")
print(news_df)
```
